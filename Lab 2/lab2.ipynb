```python
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
# Constants
S0 = 5
T = 1
mu = 0.06
sigma = 0.3
lamda = 5
```

For sampling the jumps we will use the acceptance and rejection probabilities for a jump of a possion process to occur in a discrete interval.

```python
for i in range(10):
    sample_path = [S0]
    steps = 100
    dt = T/steps
    for j in range(steps):
        
        np.random.seed(np.random.poisson(1,100))
        dS = mu*sample_path[-1]*dt + sigma*sample_path[-1]*np.sqrt(dt)*np.random.standard_normal()
        if np.random.uniform(0,1) < lamda*dt:
            # jump occurs
            dS+= abs(0.1*sample_path[-1]*np.random.standard_normal())
        sample_path.append(sample_path[-1] + dS)
    plt.plot(sample_path)
plt.title("Jump Diffusion Sample Paths")
```

For time dependent we will only to calculate the respective values of mu and sigma at each time step

```python
for i in range(10):
    sample_path = [S0]
    steps = 100
    dt = T/steps
    for j in range(steps):
        np.random.seed(np.random.poisson(1,100))
        mu = 0.0325 - 0.25*dt*j
        sigma = 0.012 + 0.0138*dt*j - 0.00125*(dt*dt*j*j)
        dS = mu*sample_path[-1]*dt + sigma*sample_path[-1]*np.sqrt(dt)*np.random.standard_normal()
        if np.random.uniform(0,1) < lamda*dt:
            # jump occurs
            dS+= abs(0.1*sample_path[-1]*np.random.standard_normal())
        sample_path.append(sample_path[-1] + dS)
    plt.plot(sample_path)
plt.title("Jump Diffusion Sample Paths")
```
